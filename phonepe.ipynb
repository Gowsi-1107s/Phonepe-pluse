```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program
import pandas as pd
import json
import os
import numpy as np
```

```python
#path to get the data as states
path_1="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path_1)
Agg_state_list
```

```python
State_list = pd.DataFrame(Agg_state_list)
State_list 
     
```

```python
#This is to extract the data's to create a dataframe
cl1={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}
for i in Agg_state_list:
    p_i=path_1+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              cl1['Transaction_type'].append(Name)
              cl1['Transaction_count'].append(count)
              cl1['Transaction_amount'].append(amount)
              cl1['State'].append(i)
              cl1['Year'].append(j)
              cl1['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(cl1)

Agg_Trans
```

```python
# Print info of DataFrame
Agg_Trans.info()
```

```python
#This is to direct the path to get the data as states

path_2="/content/pulse/data/aggregated/user/country/india/state/"
Agg_user_state_list=os.listdir(path_2)
Agg_user_state_list
```

```python
#This is to extract the data's to create a dataframe

cl2={'State':[], 'Year':[],'Quater':[],'Brands':[], 'User_count':[], 'User_percentage':[]}
for i in Agg_user_state_list:
    p_i=path_2+i+"/"
    Agg_yr=os.listdir(p_i)

    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j) 

        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            B=json.load(Data)

            try:
              for z in B['data']['usersByDevice']:
                Brand=z['brand']
                count=z['count']
                percentage=z['percentage']
                cl2['Brands'].append(Brand)
                cl2['User_count'].append(count)
                cl2['User_percentage'].append(percentage)
                cl2['State'].append(i)
                cl2['Year'].append(j)
                cl2['Quater'].append(int(k.strip('.json')))
            except:
              pass  
#Succesfully created a dataframe
Agg_User=pd.DataFrame(cl2)
     

Agg_User
```

```python
path_5="/content/pulse/data/top/transaction/country/india/state/"
Top_trans_state_list=os.listdir(path_5)
Top_trans_state_list
```
